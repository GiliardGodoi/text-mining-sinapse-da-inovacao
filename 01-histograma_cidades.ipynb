```python
import pymongo
```

```python
from util import MONGO_URI
from util import MONGO_DATABASE
```

```python
client = pymongo.MongoClient(MONGO_URI)
db = client[MONGO_DATABASE]
```

```python
db.projetos.estimated_document_count()
```

```python
group = { "$group" : {
    "_id" : "$cidade",
    "frequencia" : {"$sum" : 1}
}}

cursor = db.projetos.aggregate([group])

resultado = list()

for item in cursor:
    resultado.append(item)
    
resultado = sorted(resultado,key=lambda x: x.get('frequencia'),reverse=True)
```

```python
resultado
```

```python
data = dict()
data['cidades'] = list()
data['frequencia'] = list()
data['color'] = list()

for item in resultado[:51]:
    data['cidades'].append(item.get('_id'))
    data['frequencia'].append(item.get('frequencia'))
    color = 'steelblue' if item.get('_id') != 'Santo Antônio da Platina' else '#800000'
    data['color'].append(color)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap

output_notebook()
```

```python
source = ColumnDataSource(data=data)

grafico = figure(title='Frequência Cidades',x_range=data['cidades'],plot_width=900, plot_height=500)
grafico.vbar(x='cidades',top='frequencia',width=0.9,source=source,color='color')

grafico.xgrid.grid_line_color = None
grafico.y_range.start = 0
grafico.y_range.end = 250
# grafico.legend.orientation = "horizontal"
# grafico.legend.location = "top_center"
grafico.xaxis.major_label_orientation = 1.2
grafico.title.align = 'center'
show(grafico)
```
